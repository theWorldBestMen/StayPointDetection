```python
import skmob
```

```python
df_51 = df.loc[df['deviceid'] == 51, ['deviceid', 'datetime', 'latitude', 'longitude']]
```

```python
print(df_51.head())
```

```python
tdf_51 = skmob.TrajDataFrame(df_51, latitude='latitude', longitude='longitude', datetime='datetime', user_id = 'deviceid')
print(tdf_51.head())
```

```python
from skmob.preprocessing import detection
```

```python
stdf_51 = detection.stay_locations(tdf_51)
```

```python
print(stdf_51.head())
```

```python
print('Points of the original trajectory:\t%s'%len(tdf_51))
print('Points of stops:\t\t\t%s'%len(stdf_51))
```

```python
m = stdf_51.plot_trajectory(max_users=1, start_end_markers=False)
stdf_51.plot_stops(max_users=1, map_f=m)
```
